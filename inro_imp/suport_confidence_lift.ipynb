# CONFIDENCE

```python
is ka formuala conditional probablity vala hi h

yani 2 products ko akathi bht ziada buy hui vo bacon and eggs hn, yani agr koi bnda 
bacon buy kry to bht chances h k vo eggs bhi buy kry, but kia probability h bacon k sath eggs ko buy krni ki to vo hm
CONFIDENCE s nikalty hn
```

```python
Confidence(X => Y) = (Number of transactions containing X and Y) / (Number of transactions containing X)
Confidence(X => Y) = P(Y|X)
where X and Y are the itemsets for which you are calculating the confidence of the rule X => Y (meaning If X, then Y).
```

```python
e.g: in this dataset:
Transaction ID	Items Purchased
1	Eggs, Bread, Butter
2	Bread, Butter, Jam
3	Eggs, Milk, Bread
4	Milk, Bread, Butter, Cheese
5	Eggs, Jam

confidence of the rule {Eggs} -> {Bread} and {Milk} -> {Cheese}:
for {Eggs} -> {Bread}:
   Confidence(eggs => bread) = (Number of transactions containing eggs and bread) / (Number of transactions containing eggs)
    = Confidence for {Eggs} -> {Bread}: 2/3 or 67%
similarly:
Confidence for {Milk} -> {Cheese}: 50%
```
